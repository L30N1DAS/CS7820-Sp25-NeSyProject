```python
from api import chat_with_model, SUPPORTED_MODELS
```

```python
default_options = {
    "temperature": 0.8,
    "top_p": 0.9,
    "top_k": 40,
    "num_predict": -1,
    "presence_penalty": 0.0,
    "frequency_penalty": 0.0,
    "repeat_penalty": 1.1,
    "stop": [],
    "seed": None,
    "mirostat": 0,
    "mirostat_tau": 5.0,
    "mirostat_eta": 0.1,
    "grammar": None
}
```

```python
filled_prompt = "Whats a triplestore?"
models = ['deepseek-r1:14b', 'phi4:latest', 'gemma3:latest', 'mistral-small:22b', 'phi4-mini:latest', 'granite3.2:2b', 'llama3.2:latest']
for model in models:
    if model not in models:
        continue
    print(f"🧠 Model: {model}")
    messages = [
        {"role": "system", "content": "You are a helpful assistant and expert in knowledge graphs and knowledge engineering"},
        {"role": "user", "content": filled_prompt}
    ]
    
    try:
        response = chat_with_model(
            model_name=model,
            messages=messages,
            options=default_options
        )
        answer = response["message"]["content"].strip()
    except Exception as e:
        answer = f"Error: {e}"
    print("*" * 10)
    print(answer, end = "\n\n\n\n")
```
